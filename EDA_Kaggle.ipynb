```python
monday_start_time.dtype
```

```python
friday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 4]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Adjust figure size as needed

sns.histplot(monday_start_time.dt.hour, bins=24, kde=False, label='Monday', color='blue', alpha=0.5)
sns.histplot(sunday_start_time.dt.hour, bins=24, kde=False, label='Sunday', color='red', alpha=0.5)
sns.histplot(friday_start_time.dt.hour, bins=24, kde=False, label='Friday', color='yellow', alpha=0.5)

plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.title('Accident Start Time Distribution: Monday vs. Sunday')
plt.legend()
plt.show()
```
